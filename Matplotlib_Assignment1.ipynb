```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
df.head()
```

```python
top_5_budget = df[['original_title','budget']].sort_values('budget',ascending=False)
top_5_budget = top_5_budget.head().sort_values('budget')
top_5_budget
```

```python
plt.barh(top_5_budget.original_title,top_5_budget.budget)
plt.title('Top 5 movies budget')
plt.xlabel('budget in (billions $)')
plt.ylabel('Movies Name')
plt.show()
```

```python
years=list(range(1950,2021,10))
values=list(range(10,90,10))
```

```python
len(years)
```

```python
len(values)
```

```python
plt.plot(years,values)
plt.title('Visiting numbers over the years')
plt.ylabel('Numbers of visit')
plt.xlabel('Years')
```

## Q2: Create pie chart with equal explode values.

```python

Teams=['Al-Ahli','Al-Ettihad','Al-Hilal','Al-Nasser']
colors=['green','yellow','blue','pink']
popularity= [2500,400,550,550]

plt.pie(popularity,labels=Teams,autopct='%1.1f%%',colors=colors,startangle=270,shadow=True,radius=1,explode=[0.3,0.4,0.4,0.4])
plt.title('Popular team')
plt.show()
```

# Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
scatter_sample = df[['budget','revenue']]
scatter_sample.head()
```

```python
plt.scatter(scatter_sample.budget,scatter_sample.revenue,alpha=0.4,marker='o',s=50)
plt.title('Budget V.S revenue')
plt.xlabel('budget in (billions $)')
plt.ylabel('revenue in (billions $)')
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(scatter_sample.budget,scatter_sample.revenue,alpha=0.4,marker='o',s=50)
plt.title('Budget V.S revenue',loc='left',c='r')
plt.xlabel('budget in (billions $)',fontsize=16)
plt.ylabel('revenue in (billions $)',fontsize=16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
values_2 = list(range(110,190,10))
values_3= list(range(210,290,10))

plt.plot(years,values,label='First shop')
plt.plot(years,values_2,label='Second shop')
plt.plot(years,values_3,label='Third shop')

plt.legend(loc='upper left')

plt.title('Visiting numbers over the years')
plt.ylabel('Numbers of visit')
plt.xlabel('Years')
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize=(20,10))
plt.suptitle('Subplots')


plt.subplot(3,1,1)
plt.pie(popularity,labels=Teams,autopct='%1.1f%%',colors=colors,startangle=270,shadow=True,radius=1,explode=[0.3,0.4,0.4,0.4])
plt.title('Popular team')
plt.show()

plt.subplot(3,1,2)
plt.barh(top_5_budget.original_title,top_5_budget.budget)
plt.title('Top 5 movies budget')
plt.xlabel('budget in (billions $)')
plt.ylabel('Movies Name')
plt.show()

plt.subplot(3,1,3)
plt.plot(years,values,label='First shop')
plt.plot(years,values_2,label='Second shop')
plt.plot(years,values_3,label='Third shop')

plt.legend(loc='upper left')

plt.title('Visiting numbers over the years')
plt.ylabel('Numbers of visit')
plt.xlabel('Years')
plt.show()
```
